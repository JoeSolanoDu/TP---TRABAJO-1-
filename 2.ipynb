```python
# OBJETIVO: Indice de masa corporal
# Nombre: David Joe Solano Dueñas
# Fecha: 27/04/2025

def funcionMasaCorporal(peso, talla):
    imc = peso / (talla ** 2)

    if imc < 18.5:
        print("Usted está por debajo de su peso ideal")
    elif 18.5 <= imc <= 24.999999999:
        print("Usted se encuentra en su peso ideal")
    elif 25 <= imc <= 29.9999999999:
        print("Usted se encuentra en sobrepeso")
    elif 30 <= imc <= 34.999999999:
        print("Usted se encuentra en obesidad I")
    elif 35 <= imc <= 39.999999999:
        print("Usted se encuentra en obesidad II")
    else:
        print("Usted se encuentra en obesidad III")

    print("Su IMC es:", imc)
    return

peso = float(input("Ingrese su peso en kg: "))
talla = float(input("Ingrese su talla en m: "))

funcionMasaCorporal(peso, talla)
```
