```python
# OBJETIVO: contar Consonantes
# Nombre: David Joe Solano Dueñas
# Fecha: 27/04/2025

while True:
    menu = int(input("""
    BIENVENIDO A MI MENÚ
    1. Contar consonantes
    2. Salir
    Elija una opción: """))

    if menu == 1:
      def funcionContarConsonantes(palabra):
          palabra = palabra.upper()
          i = 0

          for letra in palabra:

              if letra in "QWRTYPSDFGHJKLZXCVBNM":
                  i += 1
          return i
      palabra = input("Ingrese una palabra: ")
      print("Número de consonantes:", funcionContarConsonantes(palabra))

    elif menu == 2:
        print("GRACIAS POR INGRESAR")
        break


    else:
        print("Opción no válida. Intente de nuevo.")
```
